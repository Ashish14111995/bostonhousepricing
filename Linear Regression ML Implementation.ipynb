# Lets load the boston house pricing Dataset

```python
boston = pd.read_csv("Boston.csv")
```

```python
boston.head()
```

```python
boston = boston.drop("Unnamed: 0",axis = 1)
```

```python
boston.head()
```

```python
boston.info()
```

```python
## Summarizing the dataset
boston.describe()
```

```python
## Check the missing value
boston.isnull().sum()
```

```python
plt.scatter(boston["crim"],boston["medv"])
plt.xlabel("Crime Rate")
plt.ylabel("Price")
```

```python
import seaborn as sns
sns.regplot(x="rm",y="medv",data = boston)
```

```python
sns.regplot(x="lstat",y="medv",data = boston)
```

```python
## Independent and Dependent features
x = boston.iloc[:,:-1]
y= boston.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
## Stadardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
## print the coefficient and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameter the model has been trained
regression.get_params()
```

```python
## Prediction with test data
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for the predication
plt.scatter(y_test,reg_pred)
```

```python
## Residuals(error)
residuals = y_test-reg_pred
```

```python
residuals
```

```python
## plot the residuals
sns.displot(residuals,kind = "kde")
```

```python
## Scatter plot with respect to predication and residuals 
## Uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

# R square and adjusted square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
# display adjusted R-squared
1- (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
